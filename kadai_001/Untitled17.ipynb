```python
# openpyxl.stylesモジュールのPatternFillクラスをインポート
from openpyxl.styles import PatternFill

# ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを変数「sheet」に格納
sheet = workbook.active

# A1セルの背景色を赤色（#FF0000）に設定
sheet['A1'].fill = PatternFill(patternType='solid', fgColor='FF0000')

workbook.save('sample4.xlsx')
```

```python
# openpyxl.stylesモジュールのFontクラスをインポート
from openpyxl.styles import Font

# ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを変数「ws」に格納
ws = workbook.active

# フォントの文字色を赤色（#FF0000）に設定
font = Font(name='Calibri', size=14, bold=True, color='FF0000')

# A1セルに文字列を入力
ws['A1'] = 'Hello, World!'

# A1セルにフォントを適用
ws['A1'].font = font

# ファイルを別名保存
workbook.save('sample5.xlsx')
```

```python
# openpyxl.stylesモジュールのBorderクラスとSideクラスをインポート
from openpyxl.styles import Border, Side

# 新規ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを取得し変数wsに格納
ws = workbook.active

# B2セルに文字列データを入力
ws['B2'] = 'Hello World!'

# Side属性で線のスタイルを指定
thin_border = Side(style='thin')

# Border属性で各辺の線のスタイルを指定
border = Border(top=thin_border, bottom=thin_border, left=thin_border, right=thin_border)

# B2のborder属性にborderを設定
ws['B2'].border = border

# ワークブックを保存
workbook.save('sample6.xlsx')
```

```python
import openpyxl

# 新規ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを変数「worksheet」に格納
worksheet = workbook.active

#A1セルに数値を入力
worksheet['A1'] = 42

# A1からB2の範囲のセルを結合する
worksheet.merge_cells('A1:B2')
# ファイルを保存する
workbook.save('sample7.xlsx')
```

```python
# ファイルを読み込む
workbook = openpyxl.load_workbook('sample7.xlsx')

# アクティブなシートを変数「sheet」に格納
worksheet = workbook.active

#A1からB2の範囲の結合されたセルを解除する
worksheet.unmerge_cells('A1:B2')

# ファイルを保存する
workbook.save('sample8.xlsx')
```

```python
import openpyxl
from openpyxl.chart import LineChart, Reference

# Workbookオブジェクトを作成する
wb = openpyxl.Workbook()

# アクティブなシートを変数「ws」に格納
ws = wb.active

# 系列データのタイトルを1行目に設定
ws.cell(row=1, column=2, value='d1')
ws.cell(row=1, column=3, value='d2')

for i in range(1, 6):
    # A2～A6 に X1～X5
    ws.cell(row=i+1, column=1, value=f"X{i}")

    # B2～B6 に 1,4,9,16,25
    ws.cell(row=i+1, column=2, value=i * i)

    # C2～C6 に 1,0,1,0,1
    ws.cell(row=i+1, column=3, value=i % 2)

# 保存
wb.save('output.xlsx')
```

```python
import openpyxl
from openpyxl.chart import LineChart, Reference

wb = openpyxl.Workbook()
ws = wb.active

ws.cell(row=1, column=2, value='d1')
ws.cell(row=1, column=3, value='d2')

for i in range(1, 6):
    ws.cell(row=i+1, column=1, value=f"X{i}")
    ws.cell(row=i+1, column=2, value=i * i)
    ws.cell(row=i+1, column=3, value=i % 2)

chart = LineChart()
chart.title = 'Sample Chart'
chart.x_axis.title = 'X'
chart.y_axis.title = 'Y'

data = Reference(ws, min_row=1, min_col=2, max_row=6, max_col=3)

chart.add_data(data, titles_from_data=True)

labels = Reference(ws, min_col=1, min_row=2, max_row=6)

chart.set_categories(labels)

chart.legend = None

ws.add_chart(chart, 'E1')

wb.save('output_LineChart.xlsx')
```

```python
import openpyxl
from openpyxl.chart import LineChart, Reference
from openpyxl.chart import BarChart, Reference

wb = openpyxl.Workbook()
ws = wb.active

ws.cell(row=1, column=2, value='d1')
ws.cell(row=1, column=3, value='d2')

for i in range(1, 6):
    ws.cell(row=i+1, column=1, value=f"X{i}")
    ws.cell(row=i+1, column=2, value=i * i)
    ws.cell(row=i+1, column=3, value=i % 2)

chart = LineChart()
chart = BarChart()
chart.title = 'Sample Chart'
chart.x_axis.title = 'X'
chart.y_axis.title = 'Y'

data = Reference(ws, min_row=1, min_col=2, max_row=6, max_col=3)

chart.add_data(data, titles_from_data=True)

labels = Reference(ws, min_col=1, min_row=2, max_row=6)

chart.set_categories(labels)

chart.legend = None

ws.add_chart(chart, 'E1')

wb.save('output_LineChart.xlsx')
wb.save('output_BarChart.xlsx')
```

```python
import openpyxl
from openpyxl.styles import Border, Side

workbook = openpyxl.Workbook()
sheet = workbook.active

header = ['商品名', '単価', '数量', '売上']
sheet.append(header)

data = [
    ['商品A', 1000, 8, 8000],
    ['商品B', 2000, 5, 10000],
    ['商品C', 500, 13, 6500]
]


for row in data:
    sheet.append(row)

thin_border = Border(left=Side(style='thin'),
                    right=Side(style='thin'),
                    top=Side(style='thin'),
                    bottom=Side(style='thin'))

for row in sheet['A1:D4']:
    for cell in row:
        cell.border = thin_border

workbook.save('売上表.xlsx')
```

```python
import openpyxl
from openpyxl.chart import BarChart, Reference

workbook = openpyxl.load_workbook('売上表.xlsx')
sheet = workbook.active

data = Reference(sheet, min_row=1, min_col=4, max_row=sheet.max_row, max_col=4)
categories_range = Reference(sheet, min_col=1, min_row=2, max_row=sheet.max_row, max_col=1)

chart = BarChart()

chart.add_data(data, titles_from_data=True)

chart.set_categories(categories_range)

chart_cell = sheet.max_row + 2


sheet.add_chart(chart, f'A{chart_cell}')

workbook.save('売上表_棒グラフ.xlsx')
```

```python
import openpyxl

workbook = openpyxl.Workbook()
sheet = workbook.active

# 請求書情報
sheet['B2'] = '請求書'
sheet['B4'] = '株式会社ABC'
sheet['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'] = '担当者名:鈴木一郎 様'

# 開始位置
start_row = 10
start_col = 2  # B列

# ヘッダー
header = ['商品', '数量', '単価', '金額']

for col_index, value in enumerate(header):
    sheet.cell(row=start_row, column=start_col + col_index, value=value)

# データ
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 5000, 5000],
]

for row_index, row_data in enumerate(data):
    for col_index, value in enumerate(row_data):
        sheet.cell(
            row=start_row + 1 + row_index,
            column=start_col + col_index,
            value=value
        )

# 保存
workbook.save('請求書.xlsx')
```

```python
import openpyxl
import datetime

workbook = openpyxl.Workbook()
sheet = workbook.active

today = datetime.date.today()

# 請求書情報
sheet['B2'] = '請求書'
sheet['B4'] = '株式会社ABC'
sheet['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'] = '担当者名:鈴木一郎 様'
sheet['F4'] = 'No.'
sheet['F5'] = '日付'
sheet['G4'] = '0001'
sheet['G5'] = today

# 明細開始位置
start_row = 10
start_col = 2  # B列

header = ['商品', '数量', '単価', '金額']

for col_index, value in enumerate(header):
    sheet.cell(row=start_row, column=start_col + col_index, value=value)

data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 5000, 5000],
    [None, None, None, 35000],
]

for row_index, row_data in enumerate(data):
    for col_index, value in enumerate(row_data):
        sheet.cell(
            row=start_row + 1 + row_index,
            column=start_col + col_index,
            value=value
        )

# 合計欄開始位置
start_row = 15
start_col = 2

data = [
    ['合計', None, None, 25000],
    ['消費税', None, None, 2500],
    ['税込合計', None, None, 27500],
]

for row_index, row_data in enumerate(data):
    for col_index, value in enumerate(row_data):
        sheet.cell(
            row=start_row + row_index,
            column=start_col + col_index,
            value=value
        )

workbook.save('請求書.xlsx')
```
