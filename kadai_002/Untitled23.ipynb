```python
import pandas as pd

# データフレームの作成
df = pd.DataFrame({
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店'],
})

# 平均値を取得
avg = df['売上'].mean()

# 平均値を列として追加（全行同じ値）
df['売上平均値'] = avg

# 判定関数
def prioritize(level):
    if level >= avg + 50:
        return 'A'
    elif level >= avg:
        return 'B'
    else:
        return 'C'

# 緊急度列を作成
df['緊急度'] = df['売上'].apply(prioritize)

# Excel出力
df.to_excel('業績.xlsx', index=False)
```
