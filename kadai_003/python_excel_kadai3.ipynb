```python
import pandas as pd
from openpyxl.styles import PatternFill

# ファイル一覧（必要に応じてフルパスに変更してください）
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

df_list = []

# ファイル読み込み
for file_name in file_list:
    temp_df = pd.read_excel(file_name)
    temp_df.columns = ['売上年', '商品売上', '金額（千円）']
    df_list.append(temp_df)

# データ結合
df = pd.concat(df_list, ignore_index=True)

# 売上年 × 商品ごとの合計
summary = df.groupby(['売上年', '商品売上'], as_index=False)['金額（千円）'].sum()

# ヘッダー色（#F2F2F2）
header_fill = PatternFill(start_color='F2F2F2',
                          end_color='F2F2F2',
                          fill_type='solid')

# Excel出力
with pd.ExcelWriter('売上集計表.xlsx', engine='openpyxl') as writer:

    # データ書き込み
    df.to_excel(writer, sheet_name='元データ', index=False)
    summary.to_excel(writer, sheet_name='年別商品合計', index=False)

    # ワークブック取得
    workbook = writer.book

    # --- 元データ ---
    sheet1 = workbook['元データ']
    for cell in sheet1[1]:  # 1行目
        cell.fill = header_fill

    # --- 集計データ ---
    sheet2 = workbook['年別商品合計']
    for cell in sheet2[1]:
        cell.fill = header_fill

print("売上集計表.xlsx を作成しました（ヘッダー色付き）")
```
